**NUMPY INDEXING AND SLICING OF** **VECTORS**

```python
arr = np.arange(11,21)
arr
```

```python
arr[6] # this shows indexing at 6th number
```

```python
arr[1:6] # this is slicing
```

```python
arr_slice =arr[3::2] # this shows slicing with steping of 2
arr_slice
```

**NUMPY INDEXING AND SLICING OF MATRIX**

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[5,4] # yahan 5 th row indexing and 4th column indexing
```

```python
slice = arr[0:2,1:3] # this shows slicing from row 0:2 and column 1:3
```

```python
slice_1=arr[3:,3:]
```

```python
arr[:,2] # pura column chahiye to : laga ke , and konsa column 2
```

**BOOLEAN** **INDEXING**

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr %2 == 0
bool_index
```

```python
arr = np.arange(11,21) # Re-initialize arr to its original state
bool_index = arr %2 == 0 # Re-create bool_index based on the re-initialized arr
arr = arr[bool_index]
arr
```
